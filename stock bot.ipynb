```python
#定义机器人和用户的输出方式，主要是用于控制态输出
bot_template = "BOT : {0}"
user_template = "USER : {0}"
```

```python
#定义与意图相对应的回复语句
responses = { 'greet': ['Hello!', 
                       'Nice to meet you!', 
                       'Hey there!', 
                       'Hey!'], 
             'goodbye': ['Goodbye!', 
                         'Bye!',
                         'See ya!',
                         'farewell!' ],
             'inquiry': ['I am a robot to help you search for stock information.', 
                         'I am a robot that can query stock information for you.'],
             'thanks': ["It's my pleasure, what else can I do for you?", 
                         'Happy to help, what else can I do for you?'],
              'goodbye': ["Bye, have a good time! :)",
                         'See you next time!  :)'],
             'default': ['Sorry, I could not understand you.'
                         'Sorry, I am not sure how to help you ']
            }

#定义查找到股票后的回复语句
responses_a = [
    "I'm sorry :( I couldn't find anything like that", 
    '{} is a great stock!', 
    '{} or {} would work!', 
    '{} is one option, but I know others too :)'
]
```

```python
#回复函数，用于判断用户消息的意图并返回相应的回复语句
def respond(message):
    # Concatenate the user's message to the end of a standard bot respone
    data = interpreter.parse(message)
    intent = data["intent"]['name']
    params = {}
    neg_params = {}
    if intent in responses:
        # Return a random matching response
        bot_message = random.choice(responses[intent])
    elif intent == "stock_search":
        response, params, neg_params= respond_a(message,params,neg_params)
        bot_message = response
    else:
        bot_message = random.choice(responses["default"])
    return bot_message
```

```python
#以下两个函数中是最开始使用在在本项目中的，之后加入了甄别否定实体技术就更新成了find_stocks(params, neg_params)函数和find_stocks(params, neg_params)函数
#寻找股票函数，在数据库中检索符合条件的股票名
def find_stock(params):
    # Create the base query
    query = 'SELECT * FROM stocks'
    # Add filter clauses for each of the parameters
    if len(params) > 0:
        filters = ["{}=?".format(k) for k in params]
        query += " WHERE " + " and ".join(filters)
    # Create the tuple of values
    t = tuple(params.values())
    # Open connection to DB
    conn = sqlite3.connect("stocks.db")
    # Create a cursor
    c = conn.cursor()
    # Execute the query
    c.execute(query, t)
    # Return the results
    return c.fetchall()

#回复函数，用于在查找到合适股票信息后回复用户，对应responses_a
def respond_c(message, params):
    # Extract the entities
    entities = interpreter.parse(message)["entities"]
    # Fill the dictionary with entities
    for ent in entities:
        params[ent["entity"]] = str(ent["value"])
    results = find_stock(params)
    names = [r[0] for r in results]
    n = min(len(results), 3)
    # Return the appropriate response
    print(responses_a[n].format(*names))
    return responses_a[n].format(*names), params
```

```python
#用于茶盅否定实体，将句子根据实体所处位置分为不同的chunk，并在每个chunk中检索是否含有not、n't。若有，则为否定实体。
def negated_ents(phrase, ent_vals):
    ents = [e for e in ent_vals if e in phrase]
    ends = sorted([phrase.index(e) + len(e) for e in ents])
    start = 0
    chunks = []
    for end in ends:
        chunks.append(phrase[start:end])
        start = end
    result = {}
    for chunk in chunks:
        for ent in ents:
            if ent in chunk:
                if "not" in chunk or "n't" in chunk:
                    result[ent] = False
                else:
                    result[ent] = True
    return result

#寻找股票函数，在数据库中检索符合条件的股票名。与之前不同的是加入了neg_params参数已实现否定实体的甄别
def find_stocks(params, neg_params):
    query = 'SELECT * FROM stocks'
    if len(params) > 0 and len(neg_params) > 0:
        filters = ["{}=?".format(k) for k in params] + ["{}!=?".format(k) for k in neg_params]
        query += " WHERE " + " and ".join(filters)
    elif len(neg_params) > 0:
        filters = ["{}!=?".format(k) for k in neg_params]
        query += " WHERE " + " and ".join(filters)
    elif len(params) > 0:
        filters = ["{}=?".format(k) for k in params]
        query += " WHERE " + " and ".join(filters)
    t = tuple(params.values())
    # open connection to DB
    conn = sqlite3.connect('stocks.db')
    # create a cursor
    c = conn.cursor()
    c.execute(query, t)
    return c.fetchall()


#回复函数，用于在查找到合适股票信息后回复用户，对应responses_a
def respond_a(message,params,neg_params):
    # Extract the entities
    entities = interpreter.parse(message)["entities"]
    ent_vals = [e["value"] for e in entities]
    # Look for negated entities
    negated = negated_ents(message, ent_vals) ###
    for ent in entities:
        if ent["value"] in negated and not negated[ent["value"]]:
            neg_params[ent["entity"]] = str(ent["value"])
        else:
            params[ent["entity"]] = str(ent["value"])
    # Find the hotels
    results = find_stocks(params, neg_params)  ###
    names = [r[0] for r in results]
    n = min(len(results),3)
    # Return the correct response
    return responses_a[n].format(*names), params, neg_params
```

```python
#解析用户信息的意图
def send_message(state, pending, message):
    data = interpreter.parse(message)
    intent = data["intent"]['name']
    # Print user_template including the user_message
    #如果用户的意图为以下几种，则状态一直为初始状态。并通过responses和responses_a得到回复消息
    if (intent =="greet") or (intent == "inquiry") or (intent == "default") or (intent == "stock_search") or (intent == "thanks") or (intent == "goodbye"):
        new_state = INIT
        pending = None
        print(user_template.format(message))
        response = respond(message)
        print(bot_template.format(response))
    #否则根据interpret(message)解析用户意图并返回下一状态的值，调用policy_rules中的相应语句进行回复
    else:
        print("USER : {}".format(message))
        new_state, response, pending_state = policy_rules[(state, interpret(message))]
        print("BOT : {}".format(response))
        if pending is not None:
            new_state, response, pending_state = policy_rules[pending]
            print("BOT : {}".format(response))        
        if pending_state is not None:
            pending = (pending_state, interpret(message))
    return response ,new_state, pending
```

```python
#定义五个状态
#·INIT=0 ：表示初始状态，用户刚与机器人对话。
#·AUTHED=1：表示认证状态，要求用户登录并输入手机号。如果不输入数字则一直保持该状态。
#·CHOOSE_INFO=2：表示选择查询信息状态，用户可在该状态下输入想要查询股票的哪一方面信息。比如latest Price、latest volume等。
#·ORDERED=3：表示已返回查找信息状态，用户可在此时选择继续查找信息状态（CHOOSE_INFO=2）或者离开状态(LEAVE)。
#·LEAVE=4：表示离开状态，用户已经结束了本次查询。

INIT=0
AUTHED=1
CHOOSE_INFO=2
ORDERED=3
LEAVE=4
global name  #用于存储需要查询股票的名字

policy_rules = {
    (INIT, "company_search"): (INIT, "you'll have to log in first, what's your phone number?", None),
    (INIT, "number"): (AUTHED, "perfect, welcome back! So now which stock would you like to know about?", None),
    (AUTHED, "company_search"): (CHOOSE_INFO, "What kind of stock information would you like to see?", None),
    (AUTHED, "info_search"): (AUTHED, "Please tell me which stock you want to know first.", None),
    (CHOOSE_INFO, "info_search"): (ORDERED, "What else can I do for you?", None),
    (CHOOSE_INFO, "company_search"): (ORDERED, "What kind of stock information would you like to see?", None),
    (ORDERED, "company_search"): (CHOOSE_INFO, "What kind of stock information would you like to see?", None),
    (INIT, "info_search"): (AUTHED, "Please tell me which stock you want to know first.",None),
    (ORDERED, "info_search"): (ORDERED, "What else can I do for you?", None),
    (ORDERED, "bye"): (LEAVE, "Bye, have a good time! :)", None),
    (CHOOSE_INFO, "bye"): (LEAVE, "Bye, have a good time! :)", None),
    (AUTHED, "bye"): (LEAVE, "Bye, have a good time! :)", None)
}



def respond_b(state, message):
    (new_state, response) = policy_rules[(state, interpret(message))]
    return new_state, response



#定义相关信息查询的回复消息格式，仅用于控制台输出
def respond_latestPrice(message):
    bot_message =  "It's latest price is " + message +"$"
    return bot_message

def respond_week52Low(message):
    bot_message =  "It's lowest price during last 52 weeks is " + message +"$"
    return bot_message

def respond_week52High(message):
    bot_message =  "It's highest price during last 52 weeks is " + message +"$"
    return bot_message

def respond_latestVolume(message):
    bot_message =  "It's lastest volume is " + message +"."
    return bot_message


    
```

```python
#用于解析用户发送消息的意图并决定是否跳转至下一状态。
def interpret(message):
    global name
    data = interpreter.parse(message)
    intent = data["intent"]['name']
    messagep=str(message)           #将字典型数据转换为字符串
    msg = messagep.lower()          #变为小写字母方便提取信息
    #如果信息中含有股票名则表明在选择想要了解的股票，返回company_search
    if( 'aapl' in msg)or('goog' in msg)or( 'tsla' in msg)or( 'sed' in msg)or( 'isrg' in msg)or( 'nvr' in msg)or( 'mkl' in msg)or( 'acn' in msg)or( 'asur' in msg):
        if 'aapl' in msg:
            name = 'aapl'
        if 'goog' in msg:
            name = 'goog'
        if 'isrg' in msg:
            name = 'isrg'
        if 'tsla' in msg:
            name = 'tsla'
        if 'sed' in msg:
            name = 'sed'
        if 'nvr' in msg:
            name = 'nvr'
        if 'mkl' in msg:
            name = 'mkl'
        if 'acn' in msg:
            name = 'acn'
        if 'asur' in msg:
            name = 'asur'
        return 'company_search'
    #如果意图为info_search则表明用户输入了想要查询股票的某一方面信息，根据信息内的关键字进行检索
    if intent == 'info_search':
        info = find_info(name)
        s = info.text
        a = json.loads(s)
        if 'latest price'in msg:
            b = a.get('latestPrice')
            b = str(b)
            response = respond_latestPrice(b)
            print(bot_template.format(response))
        if('52 weeks') in msg:
            if('lowest') in msg:
                b = a.get('week52Low')
                b = str(b)
                response = respond_week52Low(b)
                print(bot_template.format(response))
            elif('highest') in msg:
                b = a.get('week52High')
                b = str(b)
                response = respond_week52High(b)
                print(bot_template.format(response))
        if"latest volume" in msg:
            b = a.get('latestVolume')
            b = str(b)
            response = respond_latestVolume(b)
            print(bot_template.format(response))
        if"all" in msg:
            print(bot_template.format(s))
        return 'info_search'
    if intent == "goodbye":
        return "bye"
    #如果信息中包含数字，则表明用户在尝试登陆
    if any([d in msg for d in string.digits]):
        return 'number'    
    return 
```
